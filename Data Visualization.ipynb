```python
plt.figure(figsize=(10,5))
angle=np.arange(0,361,15)
rad=np.deg2rad(angle)
sin=np.sin(rad)
cos=np.cos(rad)
plt.plot(angle,sin,label='sin')
plt.plot(angle,cos,label='cos')
plt.grid()
plt.legend()
plt.xticks(range(-30,391,15))
plt.yticks([-2,-1.5,-1,-0.5,0,0.5,1,1.5,2])
plt.xlabel("Angles",fontsize=20)
plt.ylabel("Values",fontsize=20)
plt.title("sin vs cos",fontsize=30)
plt.savefig("sincos.png")
```
